```python
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

data_path = r"C:\Projects\SolarSpotting-App\storage\data.yaml"
model_save_path = r"C:\Projects\SolarSpotting-App\storage\models"
Path(model_save_path).mkdir(exist_ok=True)
```

```python
model = YOLO('yolov8s.pt')  # Load pretrained model

results = model.train(
    data=data_path,
    epochs=50,
    imgsz=512,
    batch=4,
    device='cuda' if torch.cuda.is_available() else 'cpu',
    project=model_save_path,
    name='solar_spots_yoloS_v1',
    patience=10,  # Early stopping
    save=True,
    plots=True
)
```

```python
results_dir = Path(model_save_path) / 'solar_spots_v1'
results_csv = results_dir / 'results.csv'

if results_csv.exists():
    df = pd.read_csv(results_csv)
    
    # Create 2x2 subplot
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    # Plot 1: Loss curves
    ax1.plot(df['epoch'], df['train/box_loss'], label='Train Box Loss', color='blue')
    ax1.plot(df['epoch'], df['train/cls_loss'], label='Train Class Loss', color='red')
    ax1.set_title('Training Losses')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True)
    
    # Plot 2: Validation metrics
    ax2.plot(df['epoch'], df['val/box_loss'], label='Val Box Loss', color='blue')
    ax2.plot(df['epoch'], df['val/cls_loss'], label='Val Class Loss', color='red')
    ax2.set_title('Validation Losses')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(True)
    
    # Plot 3: mAP scores
    ax3.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP@0.5', color='green')
    ax3.plot(df['epoch'], df['metrics/mAP50-95(B)'], label='mAP@0.5:0.95', color='orange')
    ax3.set_title('Mean Average Precision')
    ax3.set_xlabel('Epoch')
    ax3.set_ylabel('mAP')
    ax3.legend()
    ax3.grid(True)
    
    # Plot 4: Precision & Recall
    ax4.plot(df['epoch'], df['metrics/precision(B)'], label='Precision', color='purple')
    ax4.plot(df['epoch'], df['metrics/recall(B)'], label='Recall', color='brown')
    ax4.set_title('Precision & Recall')
    ax4.set_xlabel('Epoch')
    ax4.set_ylabel('Score')
    ax4.legend()
    ax4.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Print final metrics
    final_metrics = df.iloc[-1]
    print(f"\nFinal Training Results:")
    print(f"mAP@0.5: {final_metrics['metrics/mAP50(B)']:.3f}")
    print(f"mAP@0.5:0.95: {final_metrics['metrics/mAP50-95(B)']:.3f}")
    print(f"Precision: {final_metrics['metrics/precision(B)']:.3f}")
    print(f"Recall: {final_metrics['metrics/recall(B)']:.3f}")
```

```python
best_model_path = results_dir / 'weights' / 'best.pt'
print(f"Loading best model from: {best_model_path}")

best_model = YOLO(best_model_path)

# Test on a sample validation image
val_images = list(Path(r"C:\Projects\SolarSpotting-App\storage\images\val").glob("*.jpg"))
if val_images:
    test_image = str(val_images[0])
    print(f"Testing on: {test_image}")
    
    results = best_model(test_image)
    
    # Display the result
    import matplotlib.image as mpimg
    img = mpimg.imread(test_image)
    plt.figure(figsize=(10, 8))
    plt.imshow(img)
    plt.title(f"Test Detection on: {Path(test_image).name}")
    plt.axis('off')
    plt.show()
    
    # Show detection results
    results[0].show()

print("Training and testing complete!")
```
